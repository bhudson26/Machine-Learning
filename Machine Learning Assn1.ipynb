```python
import pandas as pd
import matplotlib.pyplot as plt
from ast import literal_eval
import numpy as np
```

```python
data = pd.read_csv("/Users/beccahudson/Desktop/goodreads_data.csv")
```

```python
df = pd.DataFrame(data=data)
df['Genres'] = df['Genres'].apply(literal_eval)
```

```python
genres = []
avg_rating = []
book = []
num_rating = []
for index, row in df.iterrows():
    avg_rating.append(row['Avg_Rating'])
    genres.append(row['Genres'])
    book.append(row['Book'])
    num_rating.append(row['Num_Ratings'])
```

```python
genres[0][1]
```

```python
search = 'Fiction'
fict_books = []
fict_rating = []
not_fictr = []
not_fictb = []
for i in range(len(df)):
    found_fiction = False
    for genre in range(len(genres[i])):
        if genres[i][genre] == search:
            fict_books.append(book[i])
            fict_rating.append(avg_rating[i])
            found_fiction = True
            break
    if not found_fiction:
        not_fictr.append(avg_rating[i])
        not_fictb.append(book[i])
```

```python
combined_dict = dict(zip(fict_books, fict_rating))
```

```python
plot1 = plt.scatter(num_rating,avg_rating)
plt.show(plot1)
```

```python
# Number of points to randomly sample
num_points_to_sample = 1000

# Generate random indices
random_indices = np.random.choice(len(num_rating), num_points_to_sample, replace=False)

# Convert random_indices to a Python list
random_indices = random_indices.tolist()

# Use random_indices to sample data from num_rating and avg_rating arrays
sampled_x = [num_rating[i] for i in random_indices]
sampled_y = [avg_rating[i] for i in random_indices]

# Plot the sampled data
plt.xticks([1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000, 11000, 12000, 13000, 14000, 15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000])
plt.scatter(sampled_x, sampled_y, label='Sampled Data')
plt.xlabel('Number of Ratings')
plt.ylabel('Average Ratings')
plt.title('Randomly Sampled Scatter Plot')
plt.legend()
plt.show()
```
